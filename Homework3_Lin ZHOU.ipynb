1. Using the mean squared error (MSE) as your objective function allows for a closed form solution to finding the maximum likelihood estimate (MLE) of your model parameters in linear regression. Let’s consider the simple, single predictor variable model $Y= \beta_0 + \beta_1 X $. 

a) Use algebra to show how you can expand out the MSE to get from i to ii below. Do your proof using Latex equation formatting.


 _i)_ $MSE(\beta_0, \beta_1)=E[ (Y-(\beta_0 + \beta_1 X))^2]$ 

**Answer:**

$$
MSE(\beta_0, \beta_1)=E[ (Y-(\beta_0 + \beta_1 X))^2] \\
=E[(Y^2-2(\beta_0 X + \beta_1 X)Y + (\beta_0 + \beta_1 X)^2 ] \\
=E[Y^2-2\beta_0 Y -2\beta_1 XY + \beta_0^2 + 2\beta_0\beta_1 X + \beta_1^2 X^2] \\
=E[Y^2] - 2\beta_0 E[Y] - 2\beta_1 E[XY] + \beta_0^2 + 2\beta_0\beta_1 E[X] + \beta_1^2 E[X^2]\\
=E[Y^2] - 2\beta_0 E[Y] - 2\beta_1 (E[XY]-E[X]E[Y])-2\beta_1E[X]E[Y]+ \beta_0^2 + 2\beta_0\beta_1 E[X] +\beta_1^2 (E[X^2]-(E[X])^2)+\beta_1^2(E[X])^2\\
=E[Y^2] - 2\beta_0 E[Y] - 2\beta_1Cov[X,Y]-2\beta_1E[X]E[Y]+ \beta_0^2 + 2\beta_0\beta_1 E[X] +\beta_1^2Var[X]+\beta_1^2(E[X])^2
$$


_ii)_ $=E[Y^2] -2 \beta_0E[Y]-2 \beta_1 Cov[X,Y]-2 \beta_1 E[X]E[Y]+ \beta_0^2 +2 \beta_0 \beta_1 E[X]+\beta_1^2 Var[X]+ \beta_1^2 (E[X])^2$

b) Prove that the MLE of $\beta_0$ is $E[Y]- \beta_1 E[X]$ by taking the derivative of _ii_ above, with respect to $\beta_0$, setting the derivative to zero, and solving for $\beta_0$.

**Answer:**


$f(\beta_0)=ii=E[Y^2] - 2\beta_0 E[Y] - 2\beta_1Cov[X,Y]-2\beta_1E[X]E[Y]+ \beta_0^2 + 2\beta_0\beta_1 E[X] +\beta_1^2Var[X]+\beta_1^2(E[X])^2$

Taking the derivative with respect to $\beta_0$ and setting the derivative to zero
$\frac {d(f(\beta_0))}{d(\beta_0)}= 0 - 2E[Y] -0 - 0 +2\beta_0 + 2\beta_1 E[X] + 0 + 0 = 0\\$
Then $\beta_0 = E[Y]-\beta_1 E[X]$



c) Prove that the MLE for $\beta_1$ is $Cov[X,Y]/Var[X]$ by taking the derivative of _ii_ above, with respect to $\beta_1$, setting the derivative to zero, and solving for $\beta_1$. 

**Answer:**

$f(\beta_1)=ii=E[Y^2] - 2\beta_0 E[Y] - 2\beta_1Cov[X,Y]-2\beta_1E[X]E[Y]+ \beta_0^2 + 2\beta_0\beta_1 E[X] +\beta_1^2Var[X]+\beta_1^2(E[X])^2$

Taking the derivative with respect to $\beta_1$ and setting the derivative to zero:

$\frac {d(f(\beta_1))}{d(\beta_1)}= 0 - 0 -2Cov[X,Y] - 2E[X]E[Y] + 0 + 2\beta_0E[X] + 2 \beta_1Var[x] + 2 \beta_1 (E[X])^2=0$

Substitute $\beta_0 with E[Y]-\beta_1E[X]$ in the above equation, then
$\frac {d(f(\beta_1))}{d(\beta_1)}= -2Cov[X,Y] - 2E[X]E[Y] + 2E[X](E[Y]-\beta_1E[X]) +  2 \beta_1Var[x] + 2 \beta_1 (E[X])^2= 0$

Then $-2Cov[X,Y]+  2 \beta_1Var[x]=0$, in other words, $\beta_1=\frac {Cov[X,Y]}{Var[x]}$

3. In the case where X and Y are standard normal variables (i.e. $N(0,1)$), show that the ordinary least squares solution converges to the same solution as the Pearson correlation coefficient $r=Cov[X,Y]/(\sigma_X * \sigma_Y)$. What exactly do they both converge to?


**Answer:**
Given that X~N(0,1), Y~N(0,1),in the ordinary least square solution,

$\hat\beta=\frac{\sum_{i=1}^n(x_i -\overline x)(y_i - \overline y)}{\sum_{i=1}^n (x_i - \overline x)^2} = \frac{\sum_{i=1}^n(x_i -\overline x)(y_i - \overline y)}1=Cov[X,Y]$

The pearson correlation coeffient 
$r = \frac {\sum_{i=1}^n(x_i - \overline x)(y_i - \overline y)}{1*1} = Cov[X,Y]$

They both converge to $Cov[X,Y]$

4. Three methods for evaluating a regression model’s performance are $r^2$, $RSE$, and the $F$ Test. Show the functional form of each method and describe conceptually what it is telling you about your model.

**Answer:**

$r^2 = \frac{TSS-RSS}{TSS}=\frac{\sum_{i=1}^n(y_i-\overline y)^2 - \sum_{i=1}^n(y_i-\hat y_i)^2}{\sum_{i=1}^n(y_i-\overline y)^2}$

$r^2$ tells us the proportion of variance that can be explained by the model.

$RSE = \sqrt{\frac{RSS}{n-p-1}}=\sqrt{\frac{\sum_{i=1}^n(y_i-\hat y_i)^2}{n-p-1}}$

The $RSE$, which is the average amount that the response derivate from the regression line, is a measure of the lack of fit of the model to the data. When the predictions obtained by the model is close to the data, the $RSE$ will be small and we can conclude that the model fits the data very well.

$F = \frac{(TSS-RSS)/p}{RSS/(n-p-1)}$

F test is used to test the null hypothesis that all of the regression coefficients are zero, i.e.,$H_0:\beta_0=\beta_1=...=\beta_p=0$;
A significant p value associated with the F test for a given n and p, tells us that at least one of regression coefficients are not zero.

6. What are high leverage observations? How do they impact your model fits? What assumption of linear regression do they violate?

**Answer:**

High leverage observations have an unusual value for the predictor variables. The unusual value of high leverage point would increase the $Var[x]$, which in turn would have a sizable impact on the estimated parameters of the model $\beta_0 (E[Y]-\beta_1 E[X])$ and $\beta_1 (\frac {Cov[X,Y]}{Var[x]})$.
It violates the assumption of homoscedasticity,i.e., the error terms have a constant variance.
